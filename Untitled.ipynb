```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("TSLA.csv")
df = df.drop(['Date'], axis=1)
```

```python
df.Close
```

```python
ma10 = df.Open.rolling(10).mean()
ma10
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = pd.DataFrame(scaler.fit_transform(df['Close'].values.reshape(-1, 1)))
scaled_data = np.array(scaled_data)

training_dataset = scaled_data[0 : int(len(scaled_data)*0.75)]
testing_dataset = scaled_data[int(len(scaled_data)*0.75) : len(scaled_data)]
```

```python
x_train = []
y_train = []
x_test = []
y_test = []

avg = 10

for i in range(avg, training_dataset.shape[0]):
    x_train.append(training_dataset[i-avg: i])
    y_train.append(training_dataset[i, 0])

    
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
for i in range(avg, testing_dataset.shape[0]):
    x_test.append(testing_dataset[i-avg: i])
    y_test.append(testing_dataset[i, 0])

x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
x_test.shape
```

```python
model = keras.Sequential([
    keras.layers.LSTM(units=50, activation='relu', input_shape=(x_train.shape[1], 1), return_sequences=True),
    keras.layers.Dropout(0.2),

    keras.layers.LSTM(units=60, activation='relu', return_sequences=True),
    keras.layers.Dropout(0.2),

    keras.layers.LSTM(units=60, activation='relu', return_sequences=True),
    keras.layers.Dropout(0.3),

    keras.layers.LSTM(units=80, activation='relu', return_sequences=True),
    keras.layers.Dropout(0.3),

    keras.layers.LSTM(units=120, activation='relu'),
    keras.layers.Dropout(0.4),

    keras.layers.Dense(units=1)
])

model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=["accuracy"]
)
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
y_pred = model.predict(x_test)
X = scaler.inverse_transform(x_test[0])
X = df.Close[0 : int(len(df)*0.75)]
Y = df.Close[int(len(df)*0.75) : len(df)]
X
```
